```python
df = pd.read_csv('cleaned_products.csv')
df = df.reset_index()
df.head()
```

```python
df = df[['product_name', 'correct_product_match', 
         'product_manufacturer', 'correct_manufacturer_match', 
         'product_category', 'correct_category_name', 
         'sub_category', 'correct_sub_category', 
         'product_type', 'correct_product_type']]
df['correct_product_type'] = np.where(df['correct_product_type'].isna(), df['product_type'], df['correct_product_type'])
df[-50:]
```

```python
iproc_product_df = pd.read_excel('data/data_v2/product_list.xlsx')
iproc_product_df['Product Name'] = iproc_product_df['Product Name'].apply(lambda x: str(x).lower().strip())
product_list = iproc_product_df['Product Name'].unique().tolist()
```

```python
ipos_products_df = df[~df['correct_product_match'].isin(product_list)]  #['correct_product_match'].to_frame()
ipos_products_df[['correct_product_match', 'correct_manufacturer_match']] = ipos_products_df[['correct_product_match', 'correct_manufacturer_match']].applymap(lambda x: str(x).capitalize())
# ipos_products_df[-50:]
```

```python
ipos_product_list = ipos_products_df[['correct_product_match', 'correct_manufacturer_match', 'correct_product_type']].drop_duplicates().reset_index(drop=True)
ipos_product_list = ipos_product_list.sort_values(by='correct_product_match').reset_index(drop=True)
ipos_product_list.to_csv('iPos_products.csv', index=False)
```
