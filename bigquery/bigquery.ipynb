```python
from google.cloud import bigquery
```

```python
def read_from_bqtable(bq_query):
    client = bigquery.Client.from_service_account_json('../Sales-Data-Cleanup/bigquery_credentials/credentials.json')
    bq_data = client.query(bq_query).to_dataframe()
    return bq_data

table_id = 'iprocure-edw.iprocure_edw.products_cleaned'
query = f"""
        SELECT *
        FROM {table_id}
        """

bq_df = read_from_bqtable(query)

bq_df.to_csv('product_details.csv', index = False)

display(bq_df.head())
```
